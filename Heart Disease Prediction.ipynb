```python
import pandas as pd
data=pd.read_csv('/content/drive/MyDrive/heart.csv')
df=pd.DataFrame(data)
print(df.head())
```

```python
x=df.loc[:,df.columns!='target']
y=df.loc[:,'target']
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
from sklearn.preprocessing import StandardScaler
ml=StandardScaler()
x_train=ml.fit_transform(x_train)
x_test=ml.transform(x_test)
print(x_train.shape)
```

```python
from keras.models import Sequential
model=Sequential()
from keras.layers import Dense
# Add the first hidden layer
model.add(Dense(256, activation='relu', input_dim=13, 
                     kernel_initializer = "uniform"))
# Add the second hidden layer
model.add(Dense(128, activation='relu', 
                     kernel_initializer = "uniform"))
# Add the third hidden layer
model.add(Dense(64, activation='relu', 
                     kernel_initializer = "uniform"))
# Add the fourth hidden laye
model.add(Dense(32, activation='relu', 
                     kernel_initializer = "uniform"))
# Add the fifth hidden layer
model.add(Dense(16, activation='relu', 
                     kernel_initializer = "uniform"))
# Add the sizth hidden layer
model.add(Dense(8, activation='relu', 
                     kernel_initializer = "uniform"))
# Add the output layer
model.add(Dense(1, activation='sigmoid', 
                     kernel_initializer = "uniform"))
```

```python
# Compile the model
model.compile(optimizer='adam',
 loss='binary_crossentropy',
 metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,batch_size=8,epochs=500)
```

```python
scores = model.evaluate(x_train, y_train)
print("Training Accuracy: %.2f%%\n" % (scores[1]*100))
scores = model.evaluate(x_test, y_test)
print("Testing Accuracy: %.2f%%\n" % (scores[1]*100))
```

```python
y_pred = model.predict(x_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python

accuracy = (cm[0][0]+cm[1][1])/(cm[0][1] + cm[1][0] +cm[0][0] +cm[1][1])
print(accuracy*100)
```
